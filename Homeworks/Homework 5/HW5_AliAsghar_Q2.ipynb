## Analysis of the data

### Age and MonthlyIncome

```python
sns.scatterplot(data = df1, x = "age", y = "monthlyincome")

correlation = df1['age'].corr(df1['monthlyincome'])
print(correlation)
```

We can see that the age and monthly income are somewhat correlated with a correlation coeffecient of around 0.5. We also see that the monthly income starts to peak at around 40 years of age which is the age at which most people are at the peak of their careers.

### Age and JobRole

```python
# plot a bar chart of jobrole and avg age
sns.barplot(data = df1, x = "jobrole", y = "age")
plt.xticks(rotation=90)
```

We can see that younger people are more likely to be in the Sales Representative role and older people are more likely to be in the Manager role. This is because the Sales Representative role is an entry level role and the Manager role is a senior level role.

### JobRole and MonthlyIncome

```python
# draw a boxplot of jobrole and age
sns.boxplot(data = df1, x = "jobrole", y = "monthlyincome")
plt.xticks(rotation=90)
```

We can see that the MonthlyIncome varies a lot with the JobRole. This is because the MonthlyIncome is directly related to the JobRole. The manager and research director roles have the highest monthly income because they are senior level roles. The Sales Representative, research scientist and laboratory technician roles have the lowest monthly income because they are entry level jobs.
