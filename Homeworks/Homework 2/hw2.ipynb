## Part II: Data Cleaning

```python
import numpy as np
```

```python
formatted_file_name = f'formatted_{file_name}.csv'
df = pd.read_csv(formatted_file_name)
```

### Empty Columns: Access Code and Email Address
Dropping the columns which are empty and deleting duplicate rows if any

```python
# new_df = df.drop(['Access Code', 'Email Address'], axis=1)
new_df = df.dropna(axis=1, how='all')
# dropping duplicate rows
new_df = new_df.drop_duplicates()
new_df.head()
```

### Fixing Age discrepancies

```python
new_df["Age (years)"] = new_df["Age (years)"].str.replace(" ", "")

idx = new_df["Age (years)"].str.len()>2
df_2 = new_df.loc[idx, "Age (years)"]
df_3 = 2023 - int(df_2)
new_df.loc[idx, "Age (years)"] = df_3
new_df.loc[:,"Age (years)"]
new_df.head()
```

### Cleaning up Country of Citizenship

```python
new_df['Country of Citizenship'] = new_df['Country of Citizenship'].str.strip()
new_df['Country of Citizenship'] = new_df['Country of Citizenship'].str.title()
new_df['Country of Citizenship'] = new_df['Country of Citizenship'].replace(['Usa', 'Us'], 'United States')
new_df['Country of Citizenship']
```

### Fixing Height discrepancies

```python
new_df['Height (Inches)'] = new_df['Height (Inches)'].str.strip()
new_df['Height (Inches)'] = new_df['Height (Inches)'].str.replace('O', '0')
new_df['Height (Inches)'] = new_df['Height (Inches)'].str.replace('l', '1')
new_df['Height (Inches)'] = new_df['Height (Inches)'].str.replace('.', '')
new_df['Height (Inches)'] = new_df['Height (Inches)'].astype(float)
new_df['Height (Inches)']
```

### Cleaning up Undergraduate Degree
Replacing similar values with the same value and removing prefixes such as 'Bachelor of' and 'BS'

```python
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].str.strip()
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].str.title()
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].replace(['Yes'], None)
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].replace(['Bit', 'B.Tech In Cs', 'Computer Applications', 'Cse'], 'Computer Science')
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].replace(['Economic', 'Economics With Finance'], 'Economics')
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].replace(['Bs Accounting', 'Accounting And Isom'], 'Accounting')
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].replace(['Ece'], 'Electrical Engineering')
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].replace(['Bachelor Of Engineering'], 'Engineering')
new_df['Undergraduate Degree'] = new_df['Undergraduate Degree'].replace(['Criminology Law And Soci'], 'Criminology')

new_df['Undergraduate Degree']
```

### Filling in missing values
Filling in missing values of numbers with the mean and the missing values of strings with the mode

```python
new_df['Height (Inches)'].fillna(new_df['Height (Inches)'].mean(), inplace=True)

column_modes = new_df.mode().iloc[0]

# Fill NaN values in each column with the corresponding mode
new_df.fillna(column_modes, inplace=True)

new_df.head()
```

### Exporting the data

```python
new_df.to_csv(f'cleaned_{file_name}.csv', index=False, encoding = 'utf-8-sig')
```
